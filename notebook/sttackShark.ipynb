```python
#adicionando coluna a partir de uma que ja existe
DF_attack_shark['teste'] = DF_attack_shark['age'] * 2
display(DF_attack_shark)
```

```python
#criar uma coluna com valor padrÃ£o
DF_attack_shark.loc[:, 'casados'] = 2
display(DF_attack_shark)
```

```python
#excluir linhas e colunas
DF_attack_shark = DF_attack_shark.drop('teste', axis=1)
DF_attack_shark = DF_attack_shark.drop(0, axis=0)
display(DF_attack_shark)
```

```python
#Delatar linhas/colunas completamente vazias
DF_attack_shark = DF_attack_shark.dropna(how='all', axis=1)
#Deletar linhas que possuem pelo menos 1 valor vazio
DF_attack_shark = DF_attack_shark.dropna()
```

```python
#preencher valores vazios
#preencher com a media da coluna
DF_attack_shark['year'] = DF_attack_shark['year'].fillna(DF_attack_shark['year'].mean())
display(DF_attack_shark)
```

```python
#preencher com ultimo valor
DF_attack_shark = DF_attack_shark.ffill()
display(DF_attack_shark)
```

```python
#calcular indicadores
ataques_por_regiao = DF_attack_shark['country'].value_counts()
display(ataques_por_regiao)
```

```python
#group by
soma_das_idades_por_regiao =  DF_attack_shark[['country', 'age']].groupby('country').sum()
display(soma_das_idades_por_regiao)
```
